```python
def second_y_1(x):
    return np.sqrt(x)

def second_y_2(x):
    return x ** 3

def second_y_3(x):
    return 2 ** x

def second_y_4(x):
    return 1/(x+1)
```

```python
plt.plot(second_y_1(x))
# plt.ylim(200,500)
plt.title('y = sqrt{x}')
```

```python
plt.plot(second_y_2(x))
# plt.ylim(200,500)
plt.title('y = x ** 3')
```

```python
plt.plot(second_y_3(x))
# plt.ylim(200,500)
plt.title('y = 2^x')
```

```python
plt.plot(second_y_4(x))
# plt.ylim(200,500)
plt.title('y = 1/(x+1)')
plt.legend()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(14,7))

#define first subplot
plt.subplot(2,2,1)
plt.plot(second_y_1(x))
# plt.ylim(200,500)
plt.title('y = \sqrt{x}')

# Define second subplot
plt.subplot(2,2,2)
plt.plot(second_y_2(x))
# plt.ylim(200,500)
plt.title('y = x^3')

plt.figure(figsize=(14,7))
#define third subplot
plt.subplot(2,2,3)
plt.plot(second_y_3(x))
# plt.ylim(200,500)
plt.title('y = 2^x')

# Define fourth subplot
plt.subplot(2,2,4)
plt.plot(second_y_4(x))
plt.grid(True, ls='--')
# plt.ylim(200,500)
plt.title('y = 1/(x+1)')
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

# these four need to be plotted on the same chart. So I can see contrast of colors on one chart

```python
plt.figure(figsize=(14,7))

#define first subplot
plt.subplot(2,2,1)
plt.plot(second_y_1(x),  c='yellow', label='x=[1,150]')
# plt.ylim(200,500)
plt.title('y = \sqrt{x}')
plt.legend()

# Define second subplot
plt.subplot(2,2,2)
plt.plot(second_y_2(x), c='red', label='x=[1,150]')
# plt.ylim(200,500)
plt.title('y = x^3')
plt.legend()

plt.figure(figsize=(14,7), label='x=[1,150]')
#define third subplot
plt.subplot(2,2,3)
plt.plot(second_y_3(x), c='blue',label='x=[1,150]')
# plt.ylim(200,500)
plt.title('y = 2^x')
plt.legend()

# Define fourth subplot
plt.subplot(2,2,4)
plt.plot(second_y_4(x),c='green', label='x=[1,150]')
# plt.ylim(200,500)
plt.title('y = 1/(x+1)')
plt.legend()
```
